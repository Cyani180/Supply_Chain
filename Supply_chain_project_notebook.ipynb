```python
import pandas as pd
import nltk
from nltk import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
#nltk.download('punkt')
#nltk.download('wordnet')
#nltk.download('punkt_tab')
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
import spacy
import re

nlp = spacy.load("de_core_news_sm")
german_stopwords = set(stopwords.words('german'))

merged_df=pd.read_csv('C:/Users/doehr/Documents/GitHub/Project_supply_chain/supply_chain_project_trustpilot_advanced_merge.csv')


def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
        text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters
        text = text.lower()  # Convert to lowercase
        words = text.split()

        return " ".join(words)
    return ""

merged_df['Comment'] = merged_df['Comment'].apply(clean_text)
merged_df['Heading'] = merged_df['Heading'].apply(clean_text)

merged_df.head()
```
